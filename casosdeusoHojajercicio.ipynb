```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_students=pd.read_csv("clean_students_complete.csv")
data_students.head()
```

```python
#CAMBIO DE INDICE

data_students.set_index("Student ID", inplace=True)
data_students
```

```python
#Se elimina la columna Unnamed
data_students=data_students.drop(["Unnamed: 0"], axis=1)
data_students
```

Presenta la forma que creas mÃ¡s conveniente para describir cuÃ¡ntos alumnos hay por grados

```python
#se realiza el conte de alumnos por grado
data_students["grade"].value_counts()
```

```python
plot=data_students["grade"].value_counts().plot(kind="bar",color='#5F9EA0',figsize=(10,5))
plt.title("Alumnos por grado")
plt.xlabel("Grado")
plt.ylabel("Alumnos")
```
